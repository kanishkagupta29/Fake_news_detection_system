```python
VOCAB_SIZE = 10000
DIMENSION = 100
MAXLEN = 20
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
def oneHot(title):
    return tf.keras.preprocessing.text.one_hot(title,VOCAB_SIZE)

news_data["title"] = news_data["title"].apply(oneHot)
```

```python
news_data.head()
```

```python
news_data[news_data['real']==0].shape
```

```python
df = pd.DataFrame(tf.keras.utils.pad_sequences(news_data["title"],padding="pre",maxlen=MAXLEN))
df.shape
```

```python
news_data = pd.concat((news_data,df),axis=1)
news_data.shape
```

```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,roc_auc_score,precision_score, confusion_matrix

# cols = [i for i in news_data.columns if (i!="index" and i!="title" and i!="real")]
cols = [i for i in range(0,20)]
seed = np.random.seed(6)

X = news_data[cols]
y = news_data[["real",'source_domain']]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20,random_state=seed)
```

```python
X_test
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Embedding(VOCAB_SIZE,DIMENSION,input_length=len(cols)))
model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128,return_sequences=True)))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128)))
model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
model.summary()
```

```python
Y_new=y_train['source_domain']
Yt_new=y_test['source_domain']
y_train.drop(['source_domain'],axis=1,inplace=True)
y_test.drop(['source_domain'],axis=1,inplace=True)
#Y_new
```

```python
earlyStopping = tf.keras.callbacks.EarlyStopping(monitor="val_accuracy",patience=3,start_from_epoch=5,restore_best_weights=True,)

history = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=20,batch_size=64,callbacks=[earlyStopping])
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download Hindi stopwords and word tokenizer
nltk.download('stopwords')
nltk.download('punkt')

# Define a function to preprocess text
def preprocess_text(text, language='english'):
    if language == 'hindi':
        # Replace with appropriate Hindi stopwords and word tokenizer
        stopwords_list = set(stopwords.words('hindi'))
        word_tokenizer = nltk.data.load('tokenizers/punkt/indian.pickle')
    else:
        stopwords_list = set(stopwords.words('english'))
        word_tokenizer = word_tokenize

    text = text.lower()
    words = word_tokenizer(text)

    # Remove stopwords
    words = [word for word in words if word not in stopwords_list]

    return ' '.join(words)
```

```python
!pip install stopwordsiso
from stopwordsiso import stopwords
import string

# Define a function to preprocess Hindi text
def preprocess_hindi_text(text):
    # Use the stopwordsiso library for Hindi stopwords
    hindi_stopwords = stopwords("hi")

    text = text.lower()
    # Remove punctuation
    text = ''.join([char for char in text if char not in string.punctuation])
    # Tokenize using whitespace
    words = text.split()
    # Remove stopwords
    words = [word for word in words if word not in hindi_stopwords]
    return ' '.join(words)

# Example usage for Hindi text
hindi_text = "यह एक उदाहरण है।"
preprocessed_hindi_text = preprocess_hindi_text(hindi_text)
print(preprocessed_hindi_text)
```
